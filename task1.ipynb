standraize value of gender :

```python
df=pd.DataFrame(data)
nRow, nCol = data.shape
print(f'{nRow} {nCol}')

df['Gender']=df['Gender'].str.lower()

df['Gender']=df['Gender'].replace({
    'm':'Male',
    'MALE':'Male',
    'male':'Male',
    'f':'Female',
    'female':'Female',
    'FEAMALE':'Female',
    'other':'unkown'
})
df.head(5)
```

 Convert date formats to a consistent type (e.g., dd-mm-yyyy)

```python
#convert str in to date
df['ScheduledDay']=pd.to_datetime(df['ScheduledDay'],utc=False)
df['AppointmentDay']=pd.to_datetime(df['AppointmentDay'],utc=False)

#assign dat formate 
df['ScheduledDay']=df['ScheduledDay'].dt.strftime('%d-%m-%Y')
df['AppointmentDay']=df['AppointmentDay'].dt.strftime('%d-%m-%Y')

#print data
df.head(5)
```
